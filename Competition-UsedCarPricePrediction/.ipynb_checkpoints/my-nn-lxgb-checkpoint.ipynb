```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error
from tensorflow import keras
from sklearn.model_selection import StratifiedKFold,KFold
```

```python
df = pd.read_csv('xgb_df.csv',index_col=0)
```

```python
df.SaleID
```

```python
print("df information:")
df.info()
df1 = df.copy()
test = df1[df1['price'].isnull()]
X_train = df1[df1['price'].notnull()].drop(['price','regDate','creatDate','SaleID','regionCode'],axis=1)
Y_train = df1[df1['price'].notnull()]['price']
X_test = df1[df1['price'].isnull()].drop(['price','regDate','creatDate','SaleID','regionCode'],axis=1)
print("test information:")
test.info()
print("X_train information:")
X_train.info()
```

```python
print(X_train.columns)
```

```python
# 特征字段
tags = X_train.columns.tolist()
# tags.remove('price')
tags
# 改进点：去掉不重要的字段保留如下字段
# 'regDate','model','brand','bodyType','fuelType','power','kilometer','notRepairedDamage','creatDate',所有v
tags = ['model', 'brand', 'bodyType', 'fuelType', 'gearbox', 'power',
       'kilometer', 'notRepairedDamage', 'v_0', 'v_1', 'v_2', 'v_3', 'v_9', 'v_11', 'v_12', 'v_13',
       'v_14',   'regDate_year',
       'regDate_month', 'regDate_day', 'creatDate_year', 'creatDate_month',
       'creatDate_day', 'car_age_day', 'car_age_year',
       'power_bin', 'model_bin','brand_count','kilometer_count','fuelType_count',
      'bodyType_count','model_count',]#'power_bin', 'model_bin','brand_count','kilometer_count','fuelType_count',
#       'bodyType_count','model_count','name_count',

# tags = ['model', 'brand', 'bodyType', 'fuelType', 'gearbox', 'power',
#        'kilometer', 'notRepairedDamage', 'v_0', 'v_1', 'v_2', 'v_3',  'v_9',  'v_11', 'v_12', 'v_13',
#        'v_14', 'power_bin', 'model_bin',  'regDate_year',
#        'regDate_month', 'regDate_day', 'creatDate_year', 'creatDate_month',
#        'creatDate_day', 'car_age_day', 'car_age_year', ]
```

```python
X_train[tags].info()
```

```python
# 特征值规范化-标准化比归一化好得多
from sklearn.preprocessing import MinMaxScaler,StandardScaler
# min_max_scaler = MinMaxScaler()
# min_max_scaler.fit(X_train[tags].values)
# x_train = min_max_scaler.transform(X_train[tags].values)
# x_test = min_max_scaler.transform(X_test[tags].values)

sds = StandardScaler()
sds.fit(X_train[tags].values)
x_train = sds.transform(X_train[tags].values)
x_test = sds.transform(X_test[tags].values)

# 获得y值
y = Y_train
```

```python
from sklearn.model_selection import train_test_split
# 数据集切分
train_x, test_x, train_y, test_y = train_test_split(x_train,y,test_size=0.2)
```

```python
# 改进点：神经元个数；减小层数epoch->减小过拟合
# 搭建模型
model = keras.Sequential([
    #全连接层：`tf.keras.layers.Dense(神经元个数，activation= "激活函数“，kernel_regularizer=哪种正则化)`
    # 输入的维度是特征的维度
    keras.layers.Dense(250, activation='relu',input_shape=[len(tags)]),
    keras.layers.Dense(250, activation='relu'),
    keras.layers.Dense(250, activation='relu'),
#     keras.layers.BatchNormalization(),
    # If you don't specify anything, no activation is applied
    keras.layers.Dense(1)
])
# 定义损失函数，优化器
model.compile(loss='mean_absolute_error',optimizer='Adam')
# 模型训练
model.fit(train_x,train_y,batch_size=500,epochs=150)
```

```python
# 模型评估（训练集，验证集）
print('训练集MAE：',mean_absolute_error(train_y, model.predict(train_x)))
print('测试集MAE：',mean_absolute_error(test_y, model.predict(test_x)))
```

```python
# 模型预测
predict_y = model.predict(x_test)
# 输出结果
result = pd.DataFrame()
result['SaleID'] = df[df['price'].isnull()].SaleID
result['price'] = predict_y
result.to_csv('./submission_nn.csv', index=False)
```

## KFold

```python
# 特征值规范化-标准化比归一化好得多
from sklearn.preprocessing import MinMaxScaler,StandardScaler

sds = StandardScaler()
sds.fit(X_train[tags].values)
x_train = sds.transform(X_train[tags].values)
x_test = sds.transform(X_test[tags].values)

# 获得y值
y_train = Y_train
```

```python
# 构造训练集和测试集
def get_train_testDF(train_df,label_df):
    skv = KFold(n_splits=10, shuffle=True, random_state=42)
    trainX = []
    trainY = []
    testX = []
    testY = []
    for train_index, test_index in skv.split(X=train_df, y=label_df):
#         train_x, train_y, test_x, test_y = train_df.iloc[train_index, :], label_df.iloc[train_index], \
#                                             train_df.iloc[test_index, :], label_df.iloc[test_index]
        train_x, train_y, test_x, test_y = train_df[train_index, :], label_df[train_index], \
                                    train_df[test_index, :], label_df[test_index]


        trainX.append(train_x)
        trainY.append(train_y)
        testX.append(test_x)
        testY.append(test_y)
        
    return trainX, testX, trainY, testY
```

```python
train_x, valid_x, train_y, valid_y  = get_train_testDF(x_train,y_train)
```

```python

pred_nn = []
for i in range(10):
    print('\n============================NN training use Data {}/10============================\n'.format(i+1))
    
    model_nn = keras.Sequential([
        #全连接层：`tf.keras.layers.Dense(神经元个数，activation= "激活函数“，kernel_regularizer=哪种正则化)`
        # 输入的维度是特征的维度
        keras.layers.Dense(250, activation='relu',input_shape=[len(tags)]),
        keras.layers.Dense(250, activation='relu'),
        keras.layers.Dense(250, activation='relu'),
        # If you don't specify anything, no activation is applied
        keras.layers.Dense(1)
    ])
    model_nn.compile(loss='mean_absolute_error',optimizer='Adam')
    model_nn.fit(train_x[i],train_y[i],batch_size=500,epochs=150)

    MAE_nn = mean_absolute_error(valid_y[i], model.predict(valid_x[i]))
    print('MAE_nn:',MAE_nn)
                                 
    pred = model_nn.predict(x_test)
    pred = pd.DataFrame(pred)
    pred_nn.append(pred)

pred_nn = pd.concat(pred_nn, axis=1)
print(pred_nn)
```

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error
from tensorflow import keras

data_train = pd.read_csv('./used_car_train_20200313.csv',sep=' ')
data_test = pd.read_csv('./used_car_testB_20200421.csv', sep=' ')
print(data_train.info())

data_train.replace(to_replace='-',value=np.nan,inplace=True)
data_test.replace(to_replace='-',value=np.nan,inplace=True)
# 中位数填充缺失值
data_train.fillna(data_train.median(),inplace=True)
data_test.fillna(data_test.median(),inplace=True)

# 特征字段
tags = data_train.columns.tolist()
tags.remove('price')
tags
# 改进点：去掉不重要的字段保留如下字段
# 'regDate','model','brand','bodyType','fuelType','power','kilometer','notRepairedDamage','creatDate',所有v

data_train[tags].info()

# 处理异常数据
data_train['power'].plot()
print(data_train['power'].describe())
# 题目要求发动机功率：范围 [ 0, 600 ]
print('*'*20)
data_train['power'][data_train['power']>600] = 600
print(data_train['power'].describe())

# 特征值规范化
min_max_scaler = MinMaxScaler()
min_max_scaler.fit(data_train[tags].values)
x_train = min_max_scaler.transform(data_train[tags].values)
x_test = min_max_scaler.transform(data_test[tags].values)

# 获得y值
y = data_train['price'].values
# 数据集切分
train_x, test_x, train_y, test_y = train_test_split(x_train,y,test_size=0.2)

# 改进点：神经元个数；减小层数epoch->减小过拟合
# 搭建模型
model = keras.Sequential([
    #全连接层：`tf.keras.layers.Dense(神经元个数，activation= "激活函数“，kernel_regularizer=哪种正则化)`
    # 输入的维度是特征的维度
    keras.layers.Dense(250, activation='relu',input_shape=[len(tags)]),
    keras.layers.Dense(250, activation='relu'),
    keras.layers.Dense(250, activation='relu'),
    # If you don't specify anything, no activation is applied
    keras.layers.Dense(1)
])
# 定义损失函数，优化器
model.compile(loss='mean_absolute_error',optimizer='Adam')
# 模型训练
model.fit(train_x,train_y,batch_size=4096,epochs=200)

# 模型评估（训练集，验证集）
print('训练集MAE：',mean_absolute_error(train_y, model.predict(train_x)))
print('测试集MAE：',mean_absolute_error(test_y, model.predict(test_x)))

# 模型预测
predict_y = model.predict(x_test)
# 输出结果
result = pd.DataFrame()
result['SaleID'] = data_test.SaleID
result['price'] = predict_y
result.to_csv('./ans_nn.csv', index=False)
```
