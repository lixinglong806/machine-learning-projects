# Curse of Dimension

```python
import numpy as np
import time

d = 10
threshold = 0.3 ** d

start_time = time.time()  

for N in range(2, 1000000000000):
    if np.log(N) / N <= threshold:
        print("d = 10, min N =", N)
        break

end_time = time.time()  
print("time consumed: %.4f s" % (end_time - start_time))
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.spatial.distance import cdist

dimensions = [1, 2, 5, 10, 20, 50, 100]
N = 1000  # number of pts
avg_dists = []
std_dists = []

for d in dimensions:
    # 1. generate N pts of d dimensions
    points = np.random.rand(N, d)

    # 2. Euclidean distance b/w pts
    dists = cdist(points, points)

    # 3. fill value in diagnal with inf(infinity) to exclude themselves
    np.fill_diagonal(dists, np.inf)

    # 4. find the shortest distance of each pt
    nearest_dists = np.min(dists, axis=1)

    # 5. record mean and std of all 'shortest distances'
    avg_dists.append(np.mean(nearest_dists))
    std_dists.append(np.std(nearest_dists))


plt.figure(figsize=(10, 5))
plt.plot(dimensions, avg_dists, 'o-', label='Average Nearest Neighbor Distance')
plt.plot(dimensions, std_dists, 's-', label='Std of Nearest Neighbor Distance')
plt.xlabel("Dimension $d$")
plt.ylabel("Distance")
plt.title("Curse of Dimensionality: Nearest Neighbor Distance")
plt.legend()
plt.grid(True)
plt.show()
```
